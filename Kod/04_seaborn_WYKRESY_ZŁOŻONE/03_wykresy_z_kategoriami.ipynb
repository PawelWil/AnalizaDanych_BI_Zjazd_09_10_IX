```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips') # tu sobie importuje te dane z bazy tips, w postaci tabeli
tips.head()
```

### Wykres z kategoriami - Striplot

```python
_ = sns.catplot(tips, x='day', y='total_bill')
```

### SwarnPlot

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='swarm')
```

### Wykresy z kategoriami - Boxplot (sporo informacji na temat danych!)

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='box')
```

### WYKRES Słupkowy - BARchart

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='bar')
```

### Ilość wystąpien - Countplot

```python
_ = sns.catplot(tips, x='day', kind='count')
```
