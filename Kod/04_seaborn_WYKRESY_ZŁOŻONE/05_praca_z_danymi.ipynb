```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# bins mówi nam o ilości słupków -zwykle się stosuje jak jest duzo opcji, żeby to zawęzić. Jak są dwie lub np. trzy opcji to tego bins się nie stosuje
# 'value_counts'=sortowanie,
# iloc[:20]=zawęzenie do np.20
#height=wysokosc,
# aspect=szerokosc
```

```python
df_house = pd.read_csv('../../Pliki/04_seaborn/house_price_filter.csv') #df_house jest naszym data frame = dt, który dalej będzie używanmy jakoźródlo danych
df_house.head()
```

```python
df_house.info()
```

### Analiza jednoczynnikowa

```python
_ = sns.catplot(df_house, x='year', kind='count')
```

```python
_ = sns.catplot(df_house, x='month', kind='count')
```

```python
# teraz wyswietlamy histogram cen
_ = sns.displot(df_house, x='price', bins=70)
```

```python
# teraz histgram dla ceny za m2
_ = sns.displot(df_house, x='price_per_meter', bins=70)
```

```python
# a teraz które piętro jest najbardziej popularne
_ = sns.catplot(df_house, x='floor', kind='count')
```

```python
# a teraz filtrujemy po powierzchni mieszkania
_ = sns.displot(df_house, x='area', bins=50) # bins mówi nam o ilości słupków -zwykle się stosuje jak jest duzo opcji, żeby to zawęzić. Jak są dwie lub np. trzy opcji to tego bins się nie stosuje
```

```python
_ = sns.catplot(df_house, x='rooms', kind='count')
```

```python
# teraz filtr pod typem oferty
```

```python
_ = sns.catplot(df_house, x='offer_type', kind='count')
```

```python
# teraz jakie najczesiec sprezdawne pod wzgledm ofert budynku
_ = sns.catplot(df_house, y='offer_type_of_building', kind='count') # ALE TU MAMY WYKRES POZIOMY - bo oparty na osi Y
```

```python
# ilosci opgłoszen w danym wojewodztwei
_ = sns.catplot(df_house, y='voivodeship', kind='count')
```

```python
#ilosc ogłoszen pod wzgledem miast, i tu zastosyujemy 'value_counts'=sortowanie, iloc[:20]=zawęzenie do np.20
_ = sns.catplot(df_house, y='city_name', kind='count', order=df_house.value_counts('city_name').iloc[:20].index)
```

### ANALIZA DWUSKŁADNIKOWA(dwuczynnikowa)

```python
#Wykres relacji
_ = sns.relplot(df_house, x='price_per_meter', y='price', height=5, aspect=3) #height=wysokosc, aspect=szerokosc
```
