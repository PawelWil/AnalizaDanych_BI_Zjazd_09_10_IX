### Import biblioteki Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Pierwszy wykres

```python
plt.plot([0,1], [0,1]) # pierwszy nawias kwadratowy to [x1, x2], zas drugi to [y1, y2]
```

```python
#pierwsza zmienna to punkt na osi x (od -1 do 1.1, o skoku 0.1, zas druga zmienna to punkty na osi y: x do potÄ™gi 2
x = np.arange(-1,1.1, 0.1)
y = x ** 2
```

```python
# teraz z tych danych robimy wykres
_ = plt.plot(x,y)
```

### TytuÅ‚ wykresy

```python
plt.plot(x,y)
_ = plt.title('Wykres x * x') # _ underscore to zmienna, do ktÃ³rej przypisujemy dane wykresu - I ZAWSZE W PONIÅ»SZYCH WYKRESACH TAK BEDZIEMY ROBIÄ† !!
```

### Dwie funkcje na jednym wykresie

```python
x1 = np.arange(-1, 1.1, 0.1) # punkty na osi x
y1 = x1 ** 3 # punkty na osi y
```

```python
plt.plot (x,y) # teraz tworzymy wykres 1-ej fukncji
_ = plt.plot(x1, y1) # teraz tworzymy wykres 2-ej fukncji
```

### Etykiety wykresÃ³w - zeby je rozrÃ³Å¼niÄ‡

```python
plt.plot (x,y, label='x^2') # label = legenda
plt.plot(x1, y1, label='x^3')
_ = plt.legend()
```

### Wiele wykresÃ³w

```python
plt.subplot(1, 2, 1) #subplot(1, 2, 1) = 1 wiersz, 2 kolumny, 1 wykres
plt.plot(x,y) # i teraz rysujemy wykres 1
plt.subplot(1,2,2) #subplot(1, 2, 1) = 1 wiersz, 2 kolumny, 2 wykres
_ = plt.plot (x1, y1) # i teraz rysujemy wykres 2
```
